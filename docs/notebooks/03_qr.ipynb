# QR
Comprehensive QR parameters.

Parameters: `array_names`, `label_color`, `label_text_color`,
`known_zero_color`, `decorators`, `strict`.

```python
import sympy as sym
import la_figures
from IPython.display import SVG, display
from matrixlayout.qr import qr_grid_svg
```

```python
A = sym.Matrix([[1, 2], [3, 4]])
W = sym.eye(2)
spec = la_figures.qr_tbl_spec(A, W, array_names=True)

svg = qr_grid_svg(**spec)
display(SVG(svg))
```

Note: QR callout labels are shifted slightly to align with arrows. For 2x2 inputs, the Q^T label uses a longer arrow to avoid overlapping R.
